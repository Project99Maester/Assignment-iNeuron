```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
data=datasets.load_iris()
transformed=decomposition.PCA(n_components=3).fit_transform(data.data)
```

```python
def convert1D(data):
    ans=[]
    for i in [0,1,2]:
        temp=[]
        for row in data:
            temp.append(row[i])
        ans.append(temp)
    return ans
transformed_data=convert1D(data.data)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.scatter3D(transformed_data[0],transformed_data[1],transformed_data[2],c=data.target);
```
